The goal of this notebook is to retrieve and stage the source datasets into the format used in the dimensional model for analysis.

This notebook generates the Economy_dimension of the weather/tourism/economy data frame.

```python
import pandas as pd
```

```python
# Define the URL for the dataset
dataset_url = 'https://raw.githubusercontent.com/noobstang/cscsi4142-project-datasets/master/data/24100043_monthlyProvincial_tourists.csv'

# Load the dataset
df = pd.read_csv(dataset_url)

# Remove apostrophes from all cells
df = df.replace({"\"": ""}, regex=True)

df.head()
```

```python
# Filter the dataset for the years 1990 to 2023 and specified GEO locations
filtered_df = df[(df['REF_DATE'].between('1990', '2023')) & df['GEO'].isin(['Canada', 'Alberta', 'British Columbia', 'Ontario', 'Quebec'])]
```

```python
# Load the date dimension
date_dimension_url = 'https://raw.githubusercontent.com/noobstang/cscsi4142-project-datasets/master/dimension/date.csv'
date_df = pd.read_csv(date_dimension_url)

# Ensure REF_DATE in filtered_df and date in date_df are in the same format
filtered_df['REF_DATE'] = pd.to_datetime(filtered_df['REF_DATE'])
date_df['date_iso'] = pd.to_datetime(date_df['date_iso'])

# Merge to link Date_key
# Note: Assuming date_df contains 'date' in YYYY-MM-DD format where day is always 01 for this operation to be accurate
filtered_df = filtered_df.merge(date_df, left_on='REF_DATE', right_on='date_iso', how='left')

```

```python
# Preparing the final DataFrame (simplified example, adjust according to your mapping and available columns)
final_df = filtered_df[['Date_key', 'GEO', 'REF_DATE']]  # Include relevant columns and mappings here
final_df['Weather_key'] = ''  # Weather_key left blank
# Add placeholders for Location_key and Economy_key, assuming mappings or further instructions will be provided
final_df['Location_key'] = ''  # Placeholder for demonstration
final_df['Economy_key'] = ''  # Placeholder for demonstration

# Specify the columns based on the task requirements
final_df.columns = ['Date_key', 'Weather_key', 'Location_key', 'Economy_key', 'Total non-resident tourists', 'United states tourists', 'Non-US foreign tourists', 'Canadian tourists returning from U.S.', 'Canadian tourists returning from abroad', 'Seasonally adjusted']

# Assuming further details will fill in the tourism-related columns and boolean for Seasonally adjusted
```
