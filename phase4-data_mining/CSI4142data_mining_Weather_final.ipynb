Data mining: data transformations for Weather dimension

For the weather_final dimension, we will perform the following transformations: <br>
1. Drop the rows that have missing values for the column "snow_v". These correspond to weather stations that did not record the snowfall amount for the location. We will not use the data from these stations.<br>
2. Drop the "location" column. We will not use the city name as part of the data mining process.<br>
3. MinMax scaler to normalize the non-key columns. (max_temperature_v, min_temperature_v, avg_temperature_v, precipitation_v, snow_v).<br>

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
weather_df = pd.read_csv("https://raw.githubusercontent.com/noobstang/cscsi4142-project-datasets/master/dimension/weather_final.csv")

# Drop rows with missing 'snow_v' values
weather_df.dropna(subset=['snow_v'], inplace=True)
```

```python
# Drop the 'location' column
weather_df.drop('location', axis=1, inplace=True)
```

```python
# Columns to scale
columns_to_scale = ['max_temperature_v', 'min_temperature_v', 'avg_temperature_v', 'precipitation_v', 'snow_v']

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Scale the columns
weather_df[columns_to_scale] = scaler.fit_transform(weather_df[columns_to_scale])

# Check the transformed dataframe
#print(weather_df.head())
```
