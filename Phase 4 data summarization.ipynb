```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import seaborn as sns
import numpy as np

# Load the weather data
weather_df = pd.read_csv('https://raw.githubusercontent.com/Kajan123/Data-Science/main/data-transforming/weather_final.csv')


# Encode the location column
label_encoder = LabelEncoder()
weather_df['location'] = label_encoder.fit_transform(weather_df['location'])

# Create a histogram for each location
for location in weather_df['location'].unique():
    plt.figure()
    plt.hist(weather_df[weather_df['location'] == location]['avg_temperature_v'], bins=20)
    plt.xlabel('Average Temperature')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Average Temperature for {label_encoder.inverse_transform([location])[0]}')
    plt.show()
```
